```python
# 建立 function 1
def say_hello():
    print("Hello World")

say_hello()
x = say_hello()
print(x)
```

```python
# 建立 function 2 有 return
def get_name():
    name = "Jarvis" # 經一連串操作取得的值
    age = 20
    return name, age

info = get_name()
n, a = get_name()
print(info)
print(n)
print(a)
```

```python
# 建立 function 3 帶參數
def get_greeting_str(name, age):

    return f"Hello! {name}!!"

result = get_greeting_str("Jarvis", 18)
result1 = get_greeting_str("Mary", None)
print(result)
print(result1)
```

```python
# 建立 function 4 預設參數值
def get_greeting(greeting="Hello", name="World"):
    return f"{greeting}, {name}"

result2 = get_greeting()
result3 = get_greeting("Good Morning", "Jarvis")
result4 = get_greeting(name="Jarvis")
print(result2)
print(result3)
print(result4)
```

```python
# SQL
import pymysql
import pymysql.cursors
def sql_query(sql, database="my_train_titanic"):
    connection = pymysql.connect(
    host="localhost",
    user="root",
    password="P@ssw0rd",
    port=3306,
    database=database,
    cursorclass=pymysql.cursors.DictCursor,   #資料要呈現dict格式
    )
    with connection.cursor() as cursor:
        cursor.execute(sql)
        result = cursor.fetchall()

    return result

my_train_titanic_tables = sql_query("SHOW TABLES;")
print(my_train_titanic_tables)
```
